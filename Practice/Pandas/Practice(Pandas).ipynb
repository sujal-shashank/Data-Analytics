# **Data Manipulation**

```python
df['popperbedroom'] = df['total_rooms'] / df['population']
df['popperbedroom']
```

```python
#Drop Columns
df = df.drop(['popperbedroom'], axis=1)
```

```python
df.columns
```

```python
#Some Numpy Concepts to add columns with random numbers
import numpy as np
df['random'] = np.random.randint(1,101,size=len(df))
```

```python
df.columns
```

```python
#Rename Columns
df = df.rename(columns = {'random':'Random_no'})
```

```python
df['Random_no']
```

```python
#Replace Values
df['Random_no'] = df['Random_no'].replace(52 , 22)
df['Random_no']
```

# **Grouping and Aggregation**

```python
#Group Data
gr_df = df.groupby('Random_no')
gr_df['latitude'].count() # how many lattitudes are assigned to same numbers
```

```python
df['total_bedrooms'].sum()
```

```python
#Aggregate Data
gr_df['total_bedrooms'].sum()
```

```python
#Group by both latitude and the Random number(Multi columns/ Multi Groups)
multi_group = df.groupby(['Random_no','latitude']).size()
multi_group
```

# **Sorting**

```python
#Sort by columns
sorted_bedrooms = df.sort_values(['total_bedrooms'],ascending=True)
sorted_bedrooms['total_bedrooms']
```

```python
df.columns
```

```python
#Sort by multiple columns
df.sort_values(['total_rooms','total_bedrooms'],ascending=[False, True])
```

**NOTE:**   

**total_rooms :** False *italicized text* (Descending): The rows with the largest number of rooms will be at the very top.

**total_bedrooms :** True (Ascending): For rows that have the same room count, the one with the smallest number of bedrooms will appear first.
